```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# df.plot

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.head()
```

```python
ts.plot()
```

# df.plot(kind='')
支持以下几种：
- bar、barh
- hist
- box
- kde、density
- area
- scatter
- hexbin
- pie

```python
ts.plot(kind='hist')
```

```python
ts.plot(kind='bar')
```
